# **Flower Iris**

Iris çiçek veri seti, İngiliz istatistikçi ve biyolog Ronald Fisher'ın 1936 tarihli makalesinde lineer diskriminant analizinin bir örneği olarak taksonomik problemlerde çoklu ölçümlerin kullanılması başlıklı makalesinde tanıtılan çok değişkenli bir veri setidir. Bazen Anderson'un Iris veri seti olarak adlandırılır çünkü Edgar Anderson, üç ilgili türün Iris çiçeklerinin morfolojik varyasyonunu ölçmek için verileri topladı. Üç türden ikisi Gaspé Yarımadası'nda “hepsi aynı meradan toplanmış ve aynı gün toplanıp aynı kişi tarafından aynı aparatla ölçülmüştür”.
Veri seti, üç Iris türünün (Iris setosa, Iris virginica ve Iris versicolor) her birinden 50 örnekten oluşmaktadır. Her numuneden dört özellik ölçüldü: çanak yaprakları ve taç yapraklarının santimetre cinsinden uzunluğu ve genişliği. Bu dört özelliğin birleşimine dayanarak Fisher, türleri birbirinden ayırmak için doğrusal bir ayırt edici model geliştirdi.

# **1. Python kütüphanelerinin import edilmesi:**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
import seaborn as sns
```

# **2. IRIS veri setinin yüklenmesi:**

```python
data = pd.read_csv("iris.data")
```

# **3. Veri setinin kolonlarını ['sepal length', 'sepal width', 'petal length', 'petal width', 'class'] şeklinde isimlendirilmesi:**

```python
data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

```python
data.columns
```

```python
print(data.head(10))
```

# **4. Her niceliksel (quantitative) özellik için ortalamasını, standart sapmasını, minimum ve maksimum değerlerini hesaplanması:**

```python
print('ortalama : ',data['sepal_length'].mean())
print('standart sapma : ',data['sepal_length'].std())
print('minumun deger : ',data['sepal_length'].min())
print('maksimum deger : ',data['sepal_length'].max())
```

```python
print('ortalama : ',data['sepal_width'].mean())
print('standart sapma : ',data['sepal_width'].std())
print('minumun deger : ', data['sepal_width'].min())
print('maksimum deger : ',data['sepal_width'].max())
```

```python
print('ortalama : ',data['petal_length'].mean())
print('standart sapma : ', data['petal_length'].std())
print('minumun deger : ', data['petal_length'].min())
print('maksimum deger : ', data['petal_length'].max())
```

```python
print('ortalama : ',data['petal_width'].mean())
print('standart sapma : ', data['petal_width'].std())
print('minumun deger : ', data['petal_width'].min())
print('maksimum deger : ', data['petal_width'].max())
```

# **8. Tüm özniteliklerin histogramlarını 8 ayrı bölmeye ayırarak ve her bölme için frekansı sayarak görüntülenmesi:**

```python
plt.figure(figsize = (10, 7))
plt.subplot(2,2,1)
x = data.sepal_length
  
plt.hist(x, bins = 8, color = "green")
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Values")
plt.show()

plt.figure(figsize = (10, 7))
plt.subplot(2,2,2)
x = data.sepal_width
  
plt.hist(x, bins = 8, color = "green")
plt.title("Sepal Width in cm")
plt.xlabel("Sepal_Width_cm")
plt.ylabel("Values")
plt.show()
  
plt.figure(figsize = (10, 7))
plt.subplot(2,2,3)
x = data.petal_length
  
plt.hist(x, bins = 8, color = "green")
plt.title("Petal Length in cm")
plt.xlabel("Petal_Length_cm")
plt.ylabel("Values")
plt.show()

plt.figure(figsize = (10, 7))
plt.subplot(2,2,4)
x = data.petal_width
  
plt.hist(x, bins = 8, color = "green")
plt.title("Petal Width in cm")
plt.xlabel("Petal_Width_cm")
plt.ylabel("Values")
plt.show()

plt.tight_layout(rect=(0,0,2,2))
```
